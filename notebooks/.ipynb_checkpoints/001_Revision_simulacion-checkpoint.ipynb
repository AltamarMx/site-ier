```python
import pandas as pd
from iertools.read import read_sql
import matplotlib.pyplot as plt
```

```python
f = "../idf/001_site.sql"
site = read_sql(f,alias=True)
data = site.data
sistemas_constructivos = site.construction_systems
site.get_construction(sistemas_constructivos)
```

```python
data.columns
```

```python
Tis = [columna for columna in data.columns if "Ti_" in columna]
coolings = [columna for columna in data.columns if "Zone Ideal" in columna]
Tis, coolings
```

```python
fig, ax = plt.subplots(2,1,figsize=(12,4),sharex=True)

for columna in Tis:
    ax[0].plot(data[columna],label=columna)

for columna in coolings:
    ax[1].plot(data[columna],label=columna)
    
ax[0].legend()
ax[1].legend()
```
